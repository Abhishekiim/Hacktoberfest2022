### get it's gray scale image

```python
plt.figure(figsize = (12, 10))
img_gray = cv2.imread("kingkohli.png")
gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = "gray")
plt.colorbar()
plt.title("Gray image of original image")
```

##### Another Way of printing gray_scale image

```python
img = PIL.Image.open("kingkohli.png")
plt.figure(figsize = (12, 10))
gray_img = img.convert("L")
plt.imshow(gray_img, cmap = "gray")
plt.colorbar()
```

###### Another Way of printing gray_scale image

```python
imag = cv2.imread("kingkohli.png", 0)
plt.figure(figsize = (12, 10))
plt.imshow(imag, cmap = "gray")
```

### histogram of the original image

```python
plt.figure(figsize = (8, 6))
plt.hist(gray.flatten(), bins=256, range=[0, 256], color = "violet")
plt.legend(["Histogram"], loc = "upper right")
plt.show()
```

```python
equ = cv2.equalizeHist(gray)
```

```python
equ
```

### the equalised image

```python
plt.figure(figsize = (12, 10))
plt.imshow(equ, cmap="gray")
```

### histogram of the equalised image

```python
plt.figure(figsize = (8, 6))
plt.hist(equ.flatten(), bins=256, range=[0, 256], color = "violet")
plt.legend(["Histogram"], loc = "upper right")
plt.show()
```

### the main plot 


- A plot containing all the images and their histograms
- there are a main image, it histogram, image's equalised inage and it's histogram
- the plot is devided into 4 subplots

```python
plt.figure(figsize=(12, 10))
a = plt.subplot(2, 2, 1)
a.set_title("Original Image")
plt.imshow(gray, cmap = "gray")

b = plt.subplot(2, 2, 2)
plt.hist(gray.flatten(), bins=256, range=[0, 256], color = "violet")
plt.legend(["Hist"], loc = "upper right")
b.set_title("Original Image Histogram")
plt.show()

plt.figure(figsize=(12, 10))
c = plt.subplot(2, 2, 3)
c.set_title("Equalised Hist Image")
plt.imshow(equ, cmap="gray")

d = plt.subplot(2, 2, 4)
plt.hist(equ.flatten(), bins=256, range=[0, 256], color = "violet")
plt.legend(["Equ_Hist"], loc = "upper right")
d.set_title("Equalised image Histogram", loc="center")
plt.show()
```
